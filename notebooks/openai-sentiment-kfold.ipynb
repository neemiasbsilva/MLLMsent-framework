```python
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/alpha3p5.csv")
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/alpha3p3.csv")
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/alpha3p2.csv")
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/alpha4p5.csv")
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/alpha4p3.csv")
df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/alpha4p2.csv")
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/p5-results.csv")
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/p3-results.csv")
# df_response = pd.read_csv("/home/neemias/PerceptSent-LLM-approach/data/gpt4-openai-only/p2-results.csv")
df_response["response"] = df_response["response"].apply(lambda x: x.replace('.', '').replace(' ', '')) # for alpha5
# df_response["response"] = df_response["sentiment"].apply(lambda x: x.replace('.', '').replace(' ', '')) # for alpha3
display(df_response.head())
display(df_response.tail())
```

```python
df = df[df["id"].isin(df_response["id"])]
df.info()
```

```python
df_metrics = pd.DataFrame([])
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_idx, val_idx) in enumerate(kfold.split(df)):
    train_df = pd.DataFrame({"text": df["text"].iloc[train_idx].to_list(), 
                                     "sentiment": df["sentiment"].iloc[train_idx].to_list()})
    val_df = pd.DataFrame({"text": df["text"].iloc[val_idx].to_list(), 
                                "sentiment": df["sentiment"].iloc[val_idx].to_list()})

    text = val_df["text"].to_list()
    target = val_df["sentiment"].to_list()

    sentiments = np.unique(target)

    if (len(sentiments) == 3):
        sent_dic = {
            "Negative": 1,
            "Neutral": 0,
            "Positive": 2,
        }
        ids = df["id"].iloc[val_idx].to_list()
        # print(len(ids))
        # display(df_response[df_response["id"] == id])
        response = [df_response[df_response["id"] == id]["response"].iloc[0] for id in ids]
    elif (len(sentiments) == 2):
        # P2+
        # sent_dic = {
        #     "Negative": 1,
        #     "Neutral": 0,
        #     "Positive": 0,
        # }
        ids = df["id"].iloc[val_idx].to_list()
        response = [df_response[df_response["id"] == id]["response"].iloc[0] for id in ids]
        # P2 -
        sent_dic = {
            "Negative": 0,
            "Neutral": 0,
            "Positive": 1,
        }
    else:
        sent_dic = {
            "Positive": 4,
            "Slightlypositive": 3,
            "Neutral": 2,
            "Slightlynegative": 1,
            "Negative": 0
        }
        ids = df["id"].iloc[val_idx].to_list()
        response = [df_response[df_response["id"] == id]["response"].iloc[0] for id in ids]
    # print(response)
    # print(target)
    pred = []
    for s in response:
        pred.append(sent_dic[s.replace(' ', '')])

    # print(pred)
    # break
    accuracy_val = accuracy_score(target, pred)
    f1_val = f1_score(target, pred, average="weighted")    
    df_metrics = pd.concat([df_metrics, pd.DataFrame({"accuracy": [accuracy_val], "f1_score": [f1_val]
                                                         })], axis=0)
df_metrics.head()
```

```python
f1_scores = df_metrics["f1_score"].to_list()
mean_f1 = np.mean(f1_scores)


# define the confidence level
confidence_level = 0.95
degrees_freedon = len(f1_scores)-1

confidence_interval = stats.t.interval(
    confidence_level, 
    degrees_freedon, 
    loc=mean_f1, 
    scale=stats.sem(f1_scores)
)

print(f"Max F1-score: {max(f1_scores)}")
print(f"Average F1-score: {mean_f1}")
print(f"Confidence interval 95%: {confidence_interval}")
print(f"Inteval: {abs(confidence_interval[0]-mean_f1)} - Interval: {abs(confidence_interval[1]-mean_f1)}")
confidence_interval = stats.t.interval(
    confidence_level, 
    degrees_freedon, 
    loc=max(f1_scores), 
    scale=stats.sem(f1_scores)
)

print(f"Inteval: {abs(confidence_interval[0]-max(f1_scores))} - Interval: {abs(confidence_interval[1]-max(f1_scores))}")
```
