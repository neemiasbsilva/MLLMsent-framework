```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_distilbert = pd.read_csv("../experiments/experiment1/logs/training_logs.csv")
df_distilbert.set_index('epoch', inplace=True)
df_distilbert.head()
```

```python
df_largebert = pd.read_csv("../experiments/experiment2/logs/training_logs.csv")
df_largebert.set_index('epoch', inplace=True)
df_largebert.head()
```

```python
df_roberta = pd.read_csv("../experiments/experiment4/logs/training_logs.csv")
df_roberta.set_index('epoch', inplace=True)
df_roberta.head()
```

```python
df_distilbert[["train_accuracy", "train_f1_score", "val_accuracy", "val_f1_score"]].plot()
# Add title
plt.title('Distil-BERT Finetuning - Accuracy/F1-score VS N. Epochs')

plt.gcf().set_size_inches(10, 6)

plt.xlabel("Epochs")
plt.savefig("../reports/distil-bert-performance.png")
plt.show()
```

```python
df_largebert[["train_accuracy", "train_f1_score", "val_accuracy", "val_f1_score"]].plot()
# Add title
plt.title('Large-BERT Finetuning - Accuracy/F1-score VS N. Epochs')

plt.gcf().set_size_inches(10, 6)

plt.xlabel("Epochs")

plt.savefig("../reports/large-bert-performance.png")

plt.show()
```

```python
df_roberta[["train_accuracy", "train_f1_score", "val_accuracy", "val_f1_score"]].plot()
# Add title
plt.title('Twitter-roBERTa-base for Sentiment Analysis Finetuning - Accuracy/F1-score VS N. Epochs')

plt.gcf().set_size_inches(10, 6)

plt.xlabel("Epochs")

plt.savefig("../reports/twitter-roBERTa-base-performance.png")

plt.show()
```
